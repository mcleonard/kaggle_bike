This notebook is going to go through some exploratory data analysis with the [bike sharing](http://www.kaggle.com/c/bike-sharing-demand) data from Kaggle. I'll do some regression analysis and other work to predict the bike sharing demand, given some data.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load data and look at the first bit
data = pd.DataFrame.from_csv('train.csv')
data.head(10)
```

```python
# Moving time and date information into the dataframe
data['ordinal'] = [each.toordinal() for each in data.index]
data['dayofweek'] = [each.dayofweek for each in data.index]
data['dayofyear'] = [each.dayofyear for each in data.index]
data['hour'] = [each.hour for each in data.index]
# Adding a column with the date in a human readable form
data['strf'] = [each.strftime("%a, %b %d, %Y") for each in data.index]
```

```python
fig, ax = plt.subplots(figsize=(10,4))
xtlabels = [gdat['strf'][0] for _, gdat in data.groupby('ordinal')]
ax.plot(data.groupby('ordinal')['registered'].agg(np.mean), label='Registered')
ax.plot(data.groupby('ordinal')['casual'].agg(np.mean), label='Casual')
ax.set_xlim(0, len(xtlabels))
ax.set_xticks(range(len(xtlabels))[::14])
ax.set_xticklabels(xtlabels[::14], rotation=90.)
ax.legend(loc='best')
_ = ax.set_ylabel("Rides per day")
fig.tight_layout()
```

Here we can see that registered and casual users have different riding patterns. Casual users ride mostly on the weekends, while registered users ride pretty much every day. Seeing this, we should model casual and registered users separately going forward. 

Registered users most likely use the bikes for commuting to work, so we should expect to see more riders in the morning and evening during rush hours. Let's look at the number of rides per hour over a week. Here we can average over day of the week and hour.

```python
reg = data.groupby(('dayofweek', 'hour'))['registered'].mean()
cas = data.groupby(('dayofweek', 'hour'))['casual'].mean()
plt.plot(reg, label='Registered')
plt.plot(cas, label='Casual')
time_markers = np.where((reg.index.labels[1]==8) + (reg.index.labels[1]==17))[0]
plt.xticks(time_markers, reg.index.labels[1][time_markers])
plt.xlabel('Hour (24H)')
plt.ylabel('Riders per hour')
_ = plt.legend(loc='best')
```

Here the ticks on the x axis show 8 AM and 5 PM. It is clear that registered users commute to work on the bikes. There is definitely an interaction between day of the week and hour of the day, so we'll want to include that in our model at some point.

Let's start building a simple linear model. We're going to need to make dummy variables for weather, season, and hour. This is easily done with pandas dummy method.

```python
data.columns
```

```python
weather_dummies = pd.get_dummies(data['weather'], prefix='weather')
season_dummies = pd.get_dummies(data['season'], prefix='season')
hour_dummies = pd.get_dummies(data['hour'], prefix='hour')
data = data.merge(weather_dummies, left_index=True, right_index=True)
data = data.merge(season_dummies, left_index=True, right_index=True)
data = data.merge(hour_dummies, left_index=True, right_index=True)
```

Rescale scalar features such as temperature and windspeed.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data['temp'] = scaler.fit_transform(data['temp'])
data['atemp'] = scaler.fit_transform(data['atemp'])
data['humidity'] = scaler.fit_transform(data['humidity'])
data['windspeed'] = scaler.fit_transform(data['windspeed'])
```

```python
features = ['temp', 'atemp', 'humidity', 
            'windspeed', 'workingday', 'holiday']
features.extend(weather_dummies.columns[1:])
features.extend(season_dummies.columns[1:])
features.extend(hour_dummies.columns[1:])
```

```python
from sklearn import linear_model
lin_model = linear_model.LinearRegression()
lin_model.fit(data[features], data['registered'])
for f, c in zip(features, lin_model.coef_):
    print('{}: {:.4}'.format(f, c))
```
